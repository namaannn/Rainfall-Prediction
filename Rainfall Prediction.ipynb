```python
x=df.drop("RainTomorrow",axis=1)
y=df["RainTomorrow"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=69,test_size=0.20,stratify=y)
```

```python
from imblearn.over_sampling import SMOTE
```

```python
sm=SMOTE(random_state=69)
x_train1,y_train1=sm.fit_resample(x_train,y_train)
```

```python
from catboost import CatBoostClassifier
```

```python
cat=CatBoostClassifier(iterations=1000,eval_metric="AUC")
cat.fit(x_train1,y_train1)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
y_pred = cat.predict(x_test)
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
from sklearn import metrics
```

```python
metrics.plot_roc_curve(cat, x_test, y_test)
metrics.roc_auc_score(y_test, y_pred, average=None)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
rfc.fit(x_train1,y_train1)
y_pred = rfc.predict(x_test)
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
